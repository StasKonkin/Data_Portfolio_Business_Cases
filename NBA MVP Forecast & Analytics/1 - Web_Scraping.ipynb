#  Collecting and processing historical NBA MVP standings data from 1991 to 2023.


```python
# Define the URL pattern for basketball reference awards page
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"

# Loop through each year
for year in years:
    # Construct the URL for the specific year
    url = url_start.format(year)
    
    # Send a request to the URL
    data = requests.get(url)
    
    # Write the HTML content to a file for the specific year
    with open("C:/Users/stask/Desktop/Break into IT/SQL learning/NBA MVP Forecast & Analytics/mvp/{}.html".format(year), 
              "w+", 
              encoding="utf-8") as f:
        f.write(data.text)
    
    # Sleep for 10 seconds to be polite to the website
    time.sleep(10)
```

```python
# Create an empty list to store DataFrames for MVP data of all years
dfs = []

# Loop through each year
for year in years:
    # Read and parse the HTML for the specific year
    with open("C:/Users/stask/Desktop/Break into IT/SQL learning/NBA MVP Forecast & Analytics/mvp/{}.html".format(year)) as f:
        page = f.read()
        soup = BeautifulSoup(page, 'html.parser')
    
    # Remove the header row from the HTML table
    soup.find('tr', class_="over_header").decompose()
    
    # Find the MVP table by its ID
    mvp_table = soup.find_all(id="mvp")[0]
    
    # Read the MVP table into a DataFrame
    mvp_df = pd.read_html(str(mvp_table))[0]
    
    # Add a 'Year' column to the DataFrame
    mvp_df["Year"] = year
    
    # Append the DataFrame to the list of DataFrames
    dfs.append(mvp_df)
```

# Scraping and organizing NBA players statistics for the seasons spanning from 1991 to 2023.

```python
dfs = []
for year in years:
    with open("C:/Users/stask/Desktop/Break into IT/SQL learning/NBA MVP Forecast & Analytics/player/{}.html".format(year)) as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()
    player_table = soup.find_all(id="per_game_stats")
    player_df = pd.read_html(str(player_table))[0]
    player_df["Year"] = year
    dfs.append(player_df)
```

# Gathering and formatting NBA team standings data from 1991 to 2023 for both Eastern and Western Conferences.

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python
for year in years:
    time.sleep(10)
    url = team_stats_url.format(year)
    
    data = requests.get(url)
    
    with open("C:/Users/stask/Desktop/Break into IT/SQL learning/NBA MVP Forecast & Analytics/team/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(data.text)
```

```python
# Create an empty list to store DataFrames for team statistics of all years
dfs = []

# Loop through each year
for year in years:
    with open("team/{}.html".format(year)) as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    
    # Remove the header row from the HTML table for Eastern Conference
    soup.find('tr', class_="thead").decompose()
    e_table = soup.find_all(id="divs_standings_E")[0]
    e_df = pd.read_html(str(e_table))[0]
    
    # Add a 'Year' column to the DataFrame
    e_df["Year"] = year
    
    # Rename the 'Eastern Conference' column to 'Team'
    e_df["Team"] = e_df["Eastern Conference"]
    del e_df["Eastern Conference"]
    
    # Append the DataFrame to the list of DataFrames
    dfs.append(e_df)
    
    # Remove the header row from the HTML table for Western Conference
    w_table = soup.find_all(id="divs_standings_W")[0]
    w_df = pd.read_html(str(w_table))[0]
    
    # Add a 'Year' column to the DataFrame
    w_df["Year"] = year
    
    # Rename the 'Western Conference' column to 'Team'
    w_df["Team"] = w_df["Western Conference"]
    del w_df["Western Conference"]
    
    # Append the DataFrame to the list of DataFrames
    dfs.append(w_df)
```

```python
teams = pd.concat(dfs)
```
